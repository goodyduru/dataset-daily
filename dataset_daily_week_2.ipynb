```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = economic_data.drop(['observation_date','total_employment','recession','manufacturers_new_orders_durable_goods',
                        'manufacturer_new_orders_consumer_goods','recession_probability', 'year', 'month'], axis=1)
y = economic_data[['recession']]
X[X.columns] = scaler.fit_transform(X[X.columns])
X.tail()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
c_space = np.logspace(1, 5, 5)
cgrid = {'C': c_space}
logr = LogisticRegression(solver='lbfgs')
logreg = GridSearchCV(logr, cgrid, scoring='neg_log_loss', cv=3)
logreg.fit(X, y)
```

```python
c_dt = {
    'max_depth': [None, 3, 4, 5],
    'min_samples_leaf': [1, 3, 5],
    'max_features': [None, 0.2, 0.4, 0.6]
}
cp = DecisionTreeClassifier()
cdt = GridSearchCV(cp, c_dt, scoring='neg_log_loss', cv=3)
cdt.fit(X, y)
```

```python
import xgboost as xgb
xc = {
    'max_depth': [3, 5],
    'n_estimators': [100, 300]
}
xcgrid = GridSearchCV(xgb.XGBClassifier(seed=432), xc, scoring='neg_log_loss', cv=3)
xcgrid.fit(X, y)
```

```python
r_dt = {
    'n_estimators': [100, 200, 300]
}
rf = RandomForestClassifier()
rft = GridSearchCV(rf, r_dt, scoring='neg_log_loss', cv=3)
rft.fit(X, y)
```

```python
all_models = pd.DataFrame({
    'Model': ['Logistic Regression', 'CART', 'Gradient Boosting', 'Random Forest'],
    'logloss': [np.abs(logreg.best_score_), np.abs(cdt.best_score_), np.abs(xcgrid.best_score_), np.abs(rft.best_score_)]
})
all_models = all_models.sort_values(by=['logloss'])
```

```python
all_models = all_models.reset_index()
all_models.drop('index', axis=1, inplace=True)
```
